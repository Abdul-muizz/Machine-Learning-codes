```python
dataset_path = '/content/drive/MyDrive/USB-IDS Datasets/PROCESSED_USB-IDS_Dataset_Latest.csv'
df = pd.read_csv(dataset_path)
```

```python
df
```

```python
df = df[df['Label'].notna()]
```

```python
columns_to_drop = ['attack_name', 'label_numerical','label_binary']
```

```python
df = df.drop(columns=columns_to_drop)
```

```python
df.columns
```

```python
features = df.drop('Label', axis=1)
target = df['Label']
```

```python
# Perform feature selection
#k_max = len(features.columns)  # Maximum number of top features to select (all features)
k_min = 10
k_max = 50
k = random.randint(1, k_max)  # Generate a random number within the range
selector = SelectKBest(score_func=f_classif, k=k)
selected_features = selector.fit_transform(features, target)
```

```python
# Get the indices of the constant features
constant_features_idx = [47, 48, 49, 62]  # Update with the correct indices from the warning

# Drop the constant features
features = features.drop(features.columns[constant_features_idx], axis=1)
```

```python
# Get the indices of the selected features
selected_indices = selector.get_support(indices=True)

# Check if any selected indices are out of bounds
valid_indices = [idx for idx in selected_indices if idx < len(features.columns)]

# Get the names of the selected features
selected_feature_names = features.columns[valid_indices]

# Print the selected feature names
print("Selected features:")
for feature in selected_feature_names:
    print(feature)
```
